# Titanic: Machine Learning from Disaster

In this guided project, the aim is to create a reproducible workflow to iterate over ideas and continue to improve the accuracy of the predictions of the survival of Titanic Passengers. 

To reach the goal, I'll create helper functions which will make feature selection, model selection/tuning and creating submissions much easier as I continue to explore the data and create new features.

# Introduction

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_selection import RFECV
from sklearn.model_selection import GridSearchCV

# Read train and cv set
train = pd.read_csv('train.csv')
holdout = pd.read_csv('test.csv')

# First 5 row of holdout
holdout.head()
```

# Preprocessing the Data

```python
# I will load functions.py that I've created before by "%load" function
# %load functions.py
def process_missing(df):
    """Handle various missing values from the data set

    Usage
    ------

    holdout = process_missing(holdout)
    """
    df["Fare"] = df["Fare"].fillna(train["Fare"].mean())
    df["Embarked"] = df["Embarked"].fillna("S")
    return df

def process_age(df):
    """Process the Age column into pre-defined 'bins' 

    Usage
    ------

    train = process_age(train)
    """
    df["Age"] = df["Age"].fillna(-0.5)
    cut_points = [-1,0,5,12,18,35,60,100]
    label_names = ["Missing","Infant","Child","Teenager","Young Adult","Adult","Senior"]
    df["Age_categories"] = pd.cut(df["Age"],cut_points,labels=label_names)
    return df

def process_fare(df):
    """Process the Fare column into pre-defined 'bins' 

    Usage
    ------

    train = process_fare(train)
    """
    cut_points = [-1,12,50,100,1000]
    label_names = ["0-12","12-50","50-100","100+"]
    df["Fare_categories"] = pd.cut(df["Fare"],cut_points,labels=label_names)
    return df

def process_cabin(df):
    """Process the Cabin column into pre-defined 'bins' 

    Usage
    ------

    train process_cabin(train)
    """
    df["Cabin_type"] = df["Cabin"].str[0]
    df["Cabin_type"] = df["Cabin_type"].fillna("Unknown")
    df = df.drop('Cabin',axis=1)
    return df

def process_titles(df):
    """Extract and categorize the title from the name column 

    Usage
    ------

    train = process_titles(train)
    """
    titles = {
        "Mr" :         "Mr",
        "Mme":         "Mrs",
        "Ms":          "Mrs",
        "Mrs" :        "Mrs",
        "Master" :     "Master",
        "Mlle":        "Miss",
        "Miss" :       "Miss",
        "Capt":        "Officer",
        "Col":         "Officer",
        "Major":       "Officer",
        "Dr":          "Officer",
        "Rev":         "Officer",
        "Jonkheer":    "Royalty",
        "Don":         "Royalty",
        "Sir" :        "Royalty",
        "Countess":    "Royalty",
        "Dona":        "Royalty",
        "Lady" :       "Royalty"
    }
    extracted_titles = df["Name"].str.extract(' ([A-Za-z]+)\.',expand=False)
    df["Title"] = extracted_titles.map(titles)
    return df

def create_dummies(df,column_name):
    """Create Dummy Columns (One Hot Encoding) from a single Column

    Usage
    ------

    train = create_dummies(train,"Age")
    """
    dummies = pd.get_dummies(df[column_name],prefix=column_name)
    df = pd.concat([df,dummies],axis=1)
    return df
```

```python
# Create a feature engineering pipeline using all the functions written above
def feature_eng_pipeline(df):
    df = process_missing(df)
    df = process_age(df)
    df = process_fare(df)
    df = process_titles(df)
    df = process_cabin(df)
    
    for col in ["Age_categories","Fare_categories",
                "Title","Cabin_type","Sex"]:
        df = create_dummies(df,col)
    return df

# Apply to train and holdout set
train = feature_eng_pipeline(train)
holdout = feature_eng_pipeline(holdout)
```

The SibSp column shows the number of siblings and/or spouses each passenger had on board, while the Parch columns shows the number of parents or children each passenger had onboard. Neither column has any missing values.

The distribution of values in both columns is skewed right, with the majority of values being zero.

I add these two columns up to explore the total number of family members each passenger had onboard. The shape of the distribution of values in this case is similar.

Looking at the survival rates of the the combined family members, I can see that few of the over 500 passengers with no family members survived, while greater numbers of passengers with family members survived, mostly the family size from 1 up to 3.

# Engineering New Features

Based on the family size distribution, I come up with a new feature which is whether a passenger is alone or not.

```python
# Create a function which decides whether a passenger is alone
def isalone(df):
    df["familysize"] = df[["SibSp","Parch"]].sum(axis=1)
    df["isalone"] = 0
    df.loc[(df["familysize"] == 0),"isalone"] = 1
    df = df.drop("familysize",axis=1)
    return df

# Apply to train set and holdout set
train = isalone(train)
holdout = isalone(holdout)
```

# Selecting the Best-Performing Features

I will use sklearn.feature_selection.RFECV to automate selecting the best-performing features using recursive feature elimination.

```python
# Create a function to select the features
def select_features(df):
    # Filter the numeric columns
    numeric_df = df.select_dtypes([np.number]).dropna(axis=1)
    
    # Select only features among numeric columns
    all_X = numeric_df.drop(['Survived', 'PassengerId'], axis=1)
    
    # Target column
    all_y = numeric_df['Survived']
    
    # Model fitting
    rfc = RandomForestClassifier(random_state=1)
    selector = RFECV(rfc, cv=10)
    selector.fit(all_X, all_y)
    
    # Filtering best columns
    best_columns = list(all_X.columns[selector.support_])
    print("Best Columns \n"+"-"*12+"\n{}\n".format(best_columns))
    return best_columns
selected_cols = select_features(train)    
```

# Selecting and Tuning Different Algorithms

In this part, I will try different algorithms with different parameters by using GridSearchCV.
The algorithms are;
- Logistic Regression
- K-Nearest Neighbors
- Random Forest

```python
# Create a function which finds the best model
def select_model(df, features):
    # Spliting all_X and all_y
    all_X = df[features]
    all_y = df['Survived']
    
    # List of dictionaries, each containing a model name,
    # it's estimator and a dict of hyperparameters
    models = [
        {
            "name": "LogisticRegression",
            "estimator": LogisticRegression(),
            "hyperparameters":
                {
                    "solver": ["newton-cg", "lbfgs", "liblinear"]
                }
        },
        {
            "name": "KNeighborsClassifier",
            "estimator": KNeighborsClassifier(),
            "hyperparameters":
                {
                    "n_neighbors": range(1,20,2),
                    "weights": ["distance", "uniform"],
                    "algorithm": ["ball_tree", "kd_tree", "brute"],
                    "p": [1,2]
                }
        },
        {
            "name": "RandomForestClassifier",
            "estimator": RandomForestClassifier(random_state=1),
            "hyperparameters":
                {
                    "n_estimators": [4, 6, 9],
                    "criterion": ["entropy", "gini"],
                    "max_depth": [2, 5, 10],
                    "max_features": ["log2", "sqrt"],
                    "min_samples_leaf": [1, 5, 8],
                    "min_samples_split": [2, 3, 5]

                }
        }
    ]
    
    # Looping the models, find the best model using gridsearchcv()
    for model in models:
        # Print model name
        print(model['name'])
        print('-'*len(model['name']))

        # Instantiate gridsearchcv()
        grid = GridSearchCV(model['estimator'], 
                            param_grid=model['hyperparameters'], 
                            cv=10)
   
        # model fitting    
        grid.fit(all_X, all_y)  
        
        # Assign the parameters and score for the best model to the dictionary. 
        model['best_params'] = grid.best_params_
        model['best_score'] = grid.best_score_
        model['best_model'] = grid.best_estimator_
        
        # Print best score and best parameters
        print("Best Score: {}".format(model["best_score"]))
        print("Best Parameters: {}\n".format(model["best_params"]))
            
    return models

# Apply model selecting
models_result = select_model(train,selected_cols)
       
```

# Making a Submission to Kaggle

To submit the predictions to kaggle, I will create a function to facilitate future submissions.

```python
# Create function to save the prediction produced by selected features and model
def save_submission_file(model,cols,filename="submission.csv"):
    # Filter the holdout set
    holdout_data = holdout[cols]
    
    # Predict the target
    predictions = model.predict(holdout_data)
    
    # Create the dataframe suitable for kaggle submission
    holdout_ids = holdout["PassengerId"]
    submission_df = {"PassengerId": holdout_ids,
                 "Survived": predictions}
    submission = pd.DataFrame(submission_df)
    
    # Save the result as csv
    submission.to_csv(filename,index=False)

# Apply the save_submission_file function
best_rf_model = models_result[2]["best_model"]
save_submission_file(best_rf_model,selected_cols)
```
